```python
#Write a Regex to validate a 10 digit mobile no.
import re
pattern = r'^[0-9]{10}$'
mobile = "9876543210"
if re.match(pattern, mobile):
    print("Valid mobile number")
else:
    print("Invalid mobile number")
```

```python
#Validate a strong password
#At least 8 characters
#One uppercase
#One lowercase
#one digit
#one special character
import re
password = "Strong@123"
pattern = r"^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@$!%*?&])[A-Za-z\d@$!%*?&]{8,}$"
if re.match(pattern, password):
    print("Valid strong password")
else:
    print("Invalid password")
```

```python
#Validate IPV4 address
import re
ip="192.168.1.1"
pattern=r'^((25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])\.){3}(25[0-5]|2[0-4][0-9]|1[0-9]{2}|[1-9]?[0-9])$'
if re.match(pattern, ip):
    print("Valid IP")
else:
    print("Invalid IP")
```

```python
#Validate IPV6 address
import re
ip="2001:0db8:85a3:0000:0000:8a2e:0370:7334"
pattern=r'^(([0-9a-fA-F]{1,4}:){7}[0-9a-fA-F]{1,4}|(::1)|(::))$'
if re.match(pattern, ip):
    print("Valid IPv6")
else:
    print("Invalid IPv6")
```

```python
# Validate Hexadecimal
import re
hex_num = "1A3F"
pattern = r'^(0x)?[0-9a-fA-F]+$'
if re.match(pattern, hex_num):
    print("Valid Hexadecimal")
else:
    print("Invalid Hexadecimal")
```

```python
# Module Project 1
# 1.Generate random numbers
# 2. Check which numbers are even/odd
# 3.Round numbers using math
# 4.Count how many times numbers appear
# 5. Saves the result in a file

import random
import math
import re
import os
from collections import Counter

numbers = [random.randint(1,10) for _ in range(10)]

# even or odd
even_numbers = []
odd_numbers = []

for num in numbers:
    num_str = str(num)
    if num % 2 == 0:
        even_numbers.append(num)
    else:
        odd_numbers.append(num)

# Calculating average
average = sum(numbers) / len(numbers)

# round off the average
rounded_avg = math.ceil(average)

# count numbers
count_numbers = Counter(numbers)

# save result using os
if not os.path.exists("easy_reports"):
    os.mkdir("easy_reports")

# file path
file_path = os.path.join("easy_reports", "number_report.txt")

# write data to a file
file = open(file_path, "w")
file.write(f"Generated Numbers: {numbers}\n")
file.write(f"Even Numbers: {even_numbers}\n")
file.write(f"Odd Numbers: {odd_numbers}\n")
file.write(f"Average (Rounded): {rounded_avg}\n")
file.write("Number Count:\n")

for num, count in count_numbers.items():
    file.write(f"{num} : {count}\n")

file.close()

print(numbers)
print(even_numbers)
print(odd_numbers)
print(rounded_avg)
print(count_numbers)
print(file_path)
```

```python
#Generate random password
#Check password strength
#Count Character used in password
#Gives a strength score using math
#Saves the result in a file using OS
# Generate random password
# Check password strength
# Count Character used in password
# Gives a strength score using math
# Saves the result in a file using OS
import random
import string
import math
import os
from collections import Counter
import re
length=10
characters=string.ascii_letters+string.digits+string.punctuation
password=''.join(random.choice(characters) for _ in range(length))
strength="Weak"
if(re.search(r"[A-Z]",password) and re.search(r"[a-z]",password) and re.search(r"[0-9]",password) and re.search(r"[!@#$%^&*(),.?\":{}|<>]",password)):
    strength="Strong"
elif(re.search(r"[A-Za-z]",password) and re.search(r"[0-9]",password)):
    strength="Medium"
char_count=Counter(password)
score=math.ceil(len(password)*len(char_count)/5)
if not os.path.exists("password_reports"):
    os.mkdir("password_reports")
file_path=os.path.join("password_reports","password_report.txt")
file=open(file_path,"w")
file.write(f"Generated Password: {password}\n")
file.write(f"Password Strength: {strength}\n")
file.write(f"Strength Score: {score}\n")
file.write("Character Count:\n")
for char,count in char_count.items():
    file.write(f"{char} : {count}\n")
file.close()
print("Password:",password)
print("Strength:",strength)
print("Score:",score)
print("Character Count:",char_count)
print("File saved at:",file_path)
```
