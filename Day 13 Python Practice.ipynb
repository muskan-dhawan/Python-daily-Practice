```python
SQLite: IT IS A LIGHTWEIGHT DATABASE THAT STORES DATA IN A SINGLE FILE AND IS INBUILT IN PYTHON


COMMANDS IN SQLite:

connect()--> OPENS DATABASE
cursor()--> RUNS SQL
execute()--> EXECUTES SQL
commit()--> SAVE CHANGES
close()--> CLOSES CONNECTION
```

```python
import sqlite3

con = sqlite3.connect("company.db")
cur = con.cursor()

cur.execute("""
CREATE TABLE IF NOT EXISTS employeesss (
    emp_id INTEGER PRIMARY KEY,
    emp_name TEXT,
    salary INTEGER
)
""")

cur.execute("INSERT OR IGNORE INTO employeesss VALUES (1, 'Rohan', 40000)")
cur.execute("INSERT OR IGNORE INTO employeesss VALUES (2, 'Riya', 50000)")

con.commit()

cur.execute("SELECT * FROM employeesss")
rows = cur.fetchall()
for i in rows:
    print(i)

con.close()
```

```python
import sqlite3

con = sqlite3.connect("company.db");
cur = con.cursor();

# DDL: Create table
cur.execute("""
CREATE TABLE IF NOT EXISTS employees (
    emp_id INTEGER PRIMARY KEY,
    emp_name TEXT,
    salary INTEGER
);
""");

# Check if column exists before ALTER
cur.execute("PRAGMA table_info(employees);");
columns = [col[1] for col in cur.fetchall()];

if "department" not in columns:
    cur.execute("""
    ALTER TABLE employees
    ADD COLUMN department TEXT;
    """);

# DML: Insert records
cur.execute("""
INSERT OR IGNORE INTO employees
(emp_id, emp_name, salary, department)
VALUES (1, 'Rohan', 40000, 'IT');
""");

cur.execute("""
INSERT OR IGNORE INTO employees
(emp_id, emp_name, salary, department)
VALUES (2, 'Riya', 50000, 'HR');
""");

# DML: Update record
cur.execute("""
UPDATE employees
SET salary = 55000
WHERE emp_id = 2;
""");

# TCL: Transaction
try:
    cur.execute("BEGIN TRANSACTION;");

    cur.execute("""
    INSERT INTO employees
    (emp_id, emp_name, salary, department)
    VALUES (3, 'Aman', 45000, 'Finance');
    """);

    cur.execute("""
    INSERT INTO employees
    (emp_id, emp_name, salary, department)
    VALUES (4, 'Neha', 48000, 'IT');
    """);

    con.commit();
except:
    con.rollback();

# DQL: Select records
cur.execute("SELECT * FROM employees;");
rows = cur.fetchall();
for row in rows:
    print(row);

# DML: Delete record
cur.execute("""
DELETE FROM employees
WHERE emp_id = 1;
""");
con.commit();

# DQL: Final select
cur.execute("SELECT * FROM employees;");
rows = cur.fetchall();
for row in rows:
    print(row);

con.close();
```
