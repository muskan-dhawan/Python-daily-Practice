```python
# Mini Project OOPS 1

class BankAccount:
    def __init__(self, acc_no, name, balance):
        self.acc_no = acc_no
        self.name = name
        self._balance = balance

    def deposit(self, amount):
        self._balance += amount
        return self._balance

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
            return self._balance
        else:
            return "Insufficient balance"

    def get_balance(self):
        return self._balance

    def calculate_interest(self):
        pass   # will be overridden


class SavingAccount(BankAccount):
    def calculate_interest(self):
        return self._balance * 0.04


class CurrentAccount(BankAccount):
    def calculate_interest(self):
        return 0


class BankApp:
    def __init__(self):
        self.account = None

    def create_account(self, acc_no, name, balance, acc_type):
        if acc_type == "savings":
            self.account = SavingAccount(acc_no, name, balance)
        else:
            self.account = CurrentAccount(acc_no, name, balance)
        return "Account created successfully"

    def deposit_money(self, amount):
        return self.account.deposit(amount)

    def withdraw_money(self, amount):
        return self.account.withdraw(amount)

    def check_balance(self):
        return self.account.get_balance()

    def get_interest(self):
        return self.account.calculate_interest()

```

```python
# Mini Project OOPS 2
#Design a console based"Employee Payroll system" using oops python/ The system should calculate salary for different loops of employees using all oops concepts.
#Employee - Abstract class
#Full time employee - child class
#part time employee - child class
#salary - encapsulated data
#payroll system - controller(HAS-A)
#output- Employee Created
#Salary - 500000
#Employee created
#Salary - 40000

class Employee:
    def __init__(self, emp_id, name):
        self.emp_id = emp_id
        self.name = name

    def calculate_salary(self):
        pass   # will be overridden


class FullTimeEmployee(Employee):
    def __init__(self, emp_id, name, monthly_salary):
        super().__init__(emp_id, name)
        self.monthly_salary = monthly_salary

    def calculate_salary(self):
        return self.monthly_salary


class PartTimeEmployee(Employee):
    def __init__(self, emp_id, name, hours_worked, rate_per_hour):
        super().__init__(emp_id, name)
        self.hours_worked = hours_worked
        self.rate_per_hour = rate_per_hour

    def calculate_salary(self):
        return self.hours_worked * self.rate_per_hour


class PayrollSystem:
    def __init__(self):
        self.employee = None

    def create_employee(self, emp_type):
        if emp_type == "fulltime":
            self.employee = FullTimeEmployee(1, "Amit", 500000)
        else:
            self.employee = PartTimeEmployee(2, "Neha", 100, 400)
        print("Employee Created")

    def get_salary(self):
        return self.employee.calculate_salary()


# -------- MAIN --------
payroll = PayrollSystem()

payroll.create_employee("fulltime")
print("Salary -", payroll.get_salary())

payroll.create_employee("parttime")
print("Salary -", payroll.get_salary())

```

```python
# Mini Project OOPS 3
# Create a Library Management System where different library items calculate borrowing charges differently.
# Library item (parent class)
# Book and magazie (child class)
# LibraryApp(main class)

# Book IS-A libraryitem
# Magazine IS-A libraryitem
# LibraryApp HAS-A libraryitem

# Output format
# Item Type: Book
# Borrow Days: 5
# Borrowing Charge: 50
# Or
# Item Type: Magazine
# Borrow Days: 3
# Borrowing Charge: 30
# Parent class

# Abstraction
class LibraryItem:
    def __init__(self, item_id, borrow_days):
        self.item_id = item_id
        self._borrow_days = borrow_days   # Encapsulation
    def get_borrow_days(self):
        return self._borrow_days
    def calculate_charge(self):
        pass
# Inheritance
class Book(LibraryItem):
    def __init__(self, item_id, borrow_days):
        super().__init__(item_id, borrow_days)
# Polymorphism
    def calculate_charge(self):
        return self._borrow_days * 10
class Magazine(LibraryItem):
    def __init__(self, item_id, borrow_days):
        super().__init__(item_id, borrow_days)
 # Polymorphism
    def calculate_charge(self):
        return self._borrow_days * 10
# HAS-A Relationship
class LibraryApp:
    def __init__(self):
        self.item = None
    def create_item(self, item_id, borrow_days, item_type):
        if item_type.lower() == "book":
            self.item = Book(item_id, borrow_days)
        else:
            self.item = Magazine(item_id, borrow_days)
    def get_item_type(self):
        return type(self.item).__name__
    def get_borrow_days(self):
        return self.item.get_borrow_days()
    def get_borrowing_charge(self):
        return self.item.calculate_charge()
app = LibraryApp()
app.create_item(1, 5, "Book")
print("Item Type:", app.get_item_type())
print("Borrow Days:", app.get_borrow_days())
print("Borrowing Charge:", app.get_borrowing_charge())
print()
app.create_item(2, 3, "magazine")
print("Item Type:", app.get_item_type())
print("Borrow Days:", app.get_borrow_days())
print("Borrowing Charge:", app.get_borrowing_charge())


```
