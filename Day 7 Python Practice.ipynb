```python
#Abstraction is mix of rule class and ready class while Interface = Rule class
#Abstraction
from abc import ABC, abstractmethod

class Payment(ABC):          # Rule + Ready class
    def start(self):         # ready method
        print("Payment started")

    @abstractmethod
    def pay(self, amount):   # rule method
        pass


class UPI(Payment):
    def pay(self, amount):
        print("Paid via UPI:", amount)


class Card(Payment):
    def pay(self, amount):
        print("Paid via Card:", amount)


obj1 = UPI()
obj1.start()
obj1.pay(500)

obj2 = Card()
obj2.start()
obj2.pay(1000)
```

```python
#Interface
from abc import ABC, abstractmethod

class PaymentInterface(ABC):   # Interface (only rules)
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def pay(self, amount):
        pass


class UPI(PaymentInterface):
    def start(self):
        print("UPI payment started")

    def pay(self, amount):
        print("UPI paid:", amount)


class Card(PaymentInterface):
    def start(self):
        print("Card payment started")

    def pay(self, amount):
        print("Card paid:", amount)


obj = UPI()
obj.start()
obj.pay(300)
```

```python
#Create a prog where an abstract class name  is course .
#it has 2 methods course_info and duration(). Then u bave to create an interface ExamInterface and it has a method exam_type
#Child class as Python_course(courdse, ExamInterface. implement 2 methods duration(), exam_type()
from abc import ABC, abstractmethod

# Abstract class (Course)
class Course:
  def course_info(self):
        print("This is a programming course")
  def duration(self):
        pass

class ExamInterface(ABC):

    @abstractmethod
    def exam_type(self):
        pass
class Python_course(Course, ExamInterface):

    def duration(self):
        print("Python course duration: 3 months")

    def exam_type(self):
        print("Exam type: Online Practical Exam")

obj = Python_course()
obj.course_info()
obj.duration()
obj.exam_type()
```

```python
# Abstract class
class Course(ABC):

    def course_info(self):
        print("This is a Java course")

    @abstractmethod
    def duration(self):
        pass


# Interface
class ExamInterface(ABC):

    @abstractmethod
    def exam_type(self):
        pass

# Child class
class Java_Course(Course, ExamInterface):

    def duration(self):
        print("Java course duration: 4 months")

    def exam_type(self):
        print("Java exam type: Offline Written Exam")


# Object
obj = Java_Course()
obj.course_info()
obj.duration()
obj.exam_type()
```

```python


# Abstract class
class Course(ABC):

    def course_info(self):
        print("This is a DSA course")

    @abstractmethod
    def duration(self):
        pass


# Interface
class ExamInterface(ABC):

    @abstractmethod
    def exam_type(self):
        pass


# Child class
class DSA_Course(Course, ExamInterface):

    def duration(self):
        print("DSA course duration: 6 months")

    def exam_type(self):
        print("DSA exam type: Online Coding Test")


# Object
obj = DSA_Course()
obj.course_info()
obj.duration()
obj.exam_type()
```

```python
# Composition example

class Engine:
    def start(self):
        print("Engine started")


class Car:
    def __init__(self, brand):
        self.brand = brand
        # Composition: Car has an Engine
        self.engine = Engine()

    def drive(self):
        print("Car Brand:", self.brand)
        # Using object of another class
        self.engine.start()


# Object creation
c = Car("Toyota")
c.drive()
```
