```python
# Import library
import math

# Square root
num = 16
result = math.sqrt(num)
print("Square root of", num, "is", result)

# Power
power_result = math.pow(2, 3)
print("2 raised to power 3 is", power_result)

# Floor and Ceil
num = 7.9
print("Floor value of", num, "is", math.floor(num))
print("Ceil value of", num, "is", math.ceil(num))

# Absolute value
abs_result = math.fabs(-5)
print("Absolute value of -5 is", abs_result)
```

```python
# Random
import random

# Dice roll
dice = random.randint(1, 6)
print(dice)

# Random number between 1 and 100
num = random.randint(1, 100)
print(num)

# Random 4-digit number
otp = random.randint(1000, 9999)
print("Generated OTP:", otp)

# Random student selection
Student = ["Rahul", "Karan", "Vishal", "Neha"]
Selected = random.choice(Student)
print("Congratulations", Selected)
```

```python
# DateTime
import datetime

# Current date and time
current = datetime.datetime.now()
print("Current Date and Time:", current)

# Today's date
Today = datetime.date.today()
print("Today's Date:", Today)

# Date, Month, Year separately
print("Current Day:", Today.day)
print("Current Month:", Today.month)
print("Current Year:", Today.year)

# Date difference
date1 = datetime.date(2026, 1, 1)
date2 = datetime.date(2026, 1, 29)

diff = date2 - date1
print("Difference between dates:", diff)
```

```python
#REGEX PATTERN
import re

text = "cat cot cut"
result = re.findall("c.t", text)
print(result)
```

```python
#Start of a string(^)
text = "Hello World"
print(bool(re.search("^Hello",text)))
print(bool(re.search("^World",text)))
```

```python
#0 or more (*)
text = "hellooo"
result= re.findall("lo*",text)
print(result)
```

```python
#one or more(+)
text = "hellooo"
result= re.findall("lo+",text)
print(result)
```

```python
#0 or 1 time(?)
text="color colour"
result=re.findall("colou?r",text)
print(result)
```

```python
#character set([])
text="apple ball cat"
result=re.findall("[abc]",text)
print(result)
```

```python
#Digits([0-9])
text="My age is 30"
result=re.findall("0-9",text)
print(result)
```

```python
#small letters([a-z])
text="My age is 30"
result=re.findall("[a-z]",text)
print(result)
```

```python
#Capital letters([A-Z])
text="My age is 30"
result=re.findall("[A-Z]",text)
print(result)
```

```python
#All alphabets([A-Z-a-z])
text="My age is 30"
result=re.findall(r"[A-z-a-z]",text)
print(result)
```

```python
#DIGITS(\d)
text="Marks: 90"
result=re.findall(r"\d",text)
print(result)
```

```python
# NON DIGITS(\D)
text="Marks: 90"
result=re.findall(r"\D",text)
print(result)
```

```python
#Words(\w)
text="Marks: 90"
result=re.findall(r"\w",text)
print(result)
```

```python

#Not word()
import re

text = "Marks: 90"
result = re.findall(r"\W", text)
print(result)
```

```python
#Space(\s)

import re

text = "Marks: 90"
result = re.findall(r"\s", text)
print(result)
```

```python
#No space(\S)
import re

text = "Marks: 90"
result = re.findall(r"\S", text)
print(result)
```

```python
#Repition Count({})
text = "text = 8999998990"
result = re.findall(r"\d{10}", text)
print(result)
```

```python
#Or operator(|)
text = "I have a cat and a dog"
result = re.findall(r"cat|dog", text)
print(result)
```

```python
#Grouping(())
text = "abab ab"
result = re.findall(r"(ab)+", text)
print(result)
```
